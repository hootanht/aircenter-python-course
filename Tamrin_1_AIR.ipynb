## Strings

<div style="direction:rtl">
برای رشته ی 'Hello' یک دستور ایندکسینگ بنویسید که کارکتر  'e' برگردانده شود.

```python
# Print out 'e' using indexing
name = "Hello"
print(name[1])
```

<div style="direction:rtl">
با متد اسلایسینگ، رشته ی مذکور را Reverse کنید.

```python
# Reverse the string using slicing
stringname = "Hello"
stringname[::-1]
```

<div style="direction:rtl">
با استفاده از دو روش ایندکسینگ (از اول به آخر و از آخر به اول) ایندکس کارکتر 'o' را نمایش دهید.

```python
# Print out the 'o'
# Method 1:
name = "Hello"
print(name[4])
```

```python
# Method 2:
stringname = "Hello"
print(stringname[::-1][0])
```

## Advanced Strings

<div style="direction:rtl">
بررسی کنید که آیا هر حرف در رشته زیر از جنس حروف کوچک است؟

```python
s = 'hello how are you Sina, are you feeling okay?'

if(s == s.lower()):
    print("The strings are lower.")
else:
    print("The strings are not lower.")
```

<div style="direction:rtl">
    کدی بنویسید که اولین ایندکسی که حرف S دارد را نمایش دهد

```python
s = 'hello how are you Sina, are you feeling okay?'

print(s.find("S"))
```

<div style="direction:rtl">
چند بار حرف "w" در رشته زیر ظاهر می شود؟

```python
s = 'twywywtwywbwhsjhwuwshshwuwwwjdjdid'

print(s.count("w"))
```

<div style="direction:rtl">
    با یک دستور پایتون قسمت های استرینگ زیر را که با خط تیره جدا شده را جدا کنید

```python
s = 'Abcd_efg_hijklmn_123'

print(s.split("_"))
```

<div style="direction:rtl">
    آدرس سایت را از ایمیل زیر چگونه جدا می کنید؟

```python
s = 'John_Miller@Tesla.com'

print(s.split("@")[1])
```

## Advanced Lists

<div style="direction:rtl">
